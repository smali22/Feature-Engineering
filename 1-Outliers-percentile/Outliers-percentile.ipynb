```python
import pandas as pd
df = pd.read_csv('heights.csv')
df.head()
```

### Detect Outliers using percentile

```python
high_threshold = df['height'].quantile(0.95)
high_threshold
```

```python
df[df['height']>high_threshold]
```

```python
low_threshold = df['height'].quantile(0.05)
low_threshold
```

```python
df[df['height']<low_threshold]
```

### Remove Outliers

```python
 df[(df['height']<high_threshold) & (df['height']>low_threshold)]
```

### Now lets explore banglore property prices dataset

```python
bhp = pd.read_csv('bhp.csv')
bhp.head()
```

```python
bhp.shape
```

```python
bhp.describe()
```

### Detect Outliers using percentile

```python
max_threshold, min_threshold = bhp['price_per_sqft'].quantile([0.001,0.999])
max_threshold, min_threshold
```

```python
bhp[bhp['price_per_sqft']>max_threshold]
```

```python
bhp[bhp['price_per_sqft']<min_threshold]
```

### Remove Outliers

```python
bhp2 = bhp[(bhp['price_per_sqft']>max_threshold) & (bhp['price_per_sqft']>max_threshold)]
bhp2
```
